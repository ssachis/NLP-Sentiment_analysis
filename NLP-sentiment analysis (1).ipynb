```python
x_train=["this was awesome, an awesome movie",
  "great movie! i liked it a lot",
  "happy ending!awesome acting by the hero",
  "loved it! truly great",
  "bad not upto the mark",
  "could have been better",
  "surely a dissapointing movie"]
y_train=[1,1,1,1,0,0,0]
#1->good
#0->bad
x_test=["i was happy and i loved the acting in the movie","the movie i saw was bad"]
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("stopwords")
```

```python
tokenizer= RegexpTokenizer(r'\w+')
en_stopwords=set(stopwords.words('english'))
ps=PorterStemmer()
```

```python
def getCleanText(text):
    text=text.lower()
    tokens=tokenizer.tokenize(text)
    new_tokens=[token for token in tokens if token not in en_stopwords]#tokenization ans stopword removal
    
    stemmed_tokens=[ps.stem(token) for token in new_tokens]
    clean_text=" ".join(stemmed_tokens)
    return clean_text
```

```python
def c(x):
    x_clean=[]
    for i in x:
        x_clean.append(getCleanText(i))
    return x_clean
```

```python
x_train=c(x_train)
```

```python
x_train
```

```python
x_test=c(x_test)
```

```python
x_test
```

```python
from sklearn .feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
x_vec=cv.fit_transform(x_train).toarray()
```
